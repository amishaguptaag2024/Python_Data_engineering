```python
sorting_key = input("Enter the column name to sort the data: ")
```

```python
def _safe_cast(x):
    try:
        return int(x)
    except:
        try:
            return float(x)
        except:
            return x

cols = schema_dict.get('orders', [])
col_names = [c['column_name'] for c in cols]

if sorting_key in col_names:
    idx = col_names.index(sorting_key)
    # sort the orders (use orders_list or `order` if you meant the sample)
    orders_sorted = sorted(orders_list, key=lambda r: _safe_cast(r.split(',')[idx]))
    # optional: show first 10
    orders_sorted[:10]
else:
    # fallback: sort schema columns by their 'column_name'
    columns = sorted(cols, key=lambda c: c['column_name'])
    columns
```
