```python
# Joining Data in Pandas Dataframes
# Here are the steps involved in joining data using Pandas Dataframe APIs.

# First, we need to define index for both the dataframes that are supposed to be joined.
# Use join to join the Dataframes based on index. The default join is left, we need to make sure right join is used based on the requirement.
# In our case, we will join orders and customers using inner join.

import pandas as pd
```

```python
import json
```

```python
schemas = json.load(open('/Users/amisha/Desktop/Data_engineering/data/retail_db/schemas.json'))
```

```python
def get_column_names(schemas,ds_name,sorting_key='column_position'):
    column_details = schemas[ds_name]
    column = sorted(column_details, key=lambda x: x[sorting_key])
    return [col['column_name'] for col in column]
```

```python
order_columns = get_column_names(schemas,'orders')
```

```python
order_columns
```

```python
orders = pd.read_csv('/Users/amisha/Desktop/Data_engineering/data/retail_db/orders/part-00000',
    header=None,
    names=order_columns)
```

```python
orders
```

```python
customer_columns = get_column_names(schemas,'customers')
```

```python
customer_columns
```

```python
customers = pd.read_csv('/Users/amisha/Desktop/Data_engineering/data/retail_db/customers/part-00000',
    header=None,
    names=customer_columns)
```

```python
customers
```

```python
customers.set_index('customer_id', inplace=True)
```

```python
orders.set_index('order_customer_id', inplace=True)
```

```python
customers_orders = customers.join(orders, how='inner')
```

```python
customers_orders
```

```python
customers_orders. \
    reset_index(names='customer_id'). \
    groupby('customer_id')['customer_id']. \
    agg(order_count='count'). \
    reset_index(). \
    query('order_count >= 10')
```
